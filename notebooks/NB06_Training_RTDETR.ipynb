## Status: Deferred to Future Work

RT-DETR-l training has been **deferred** for the following reasons:

### 1. YOLO11s results already exceed project requirements

| Metric | YOLO11s (best.pt) |
|---|---|
| Test mAP50 | **93.2%** |
| Test mAP50-95 | **63.9%** |
| vest AP50 (test) | **91.9%** |
| no-vest AP50 (test) | **92.1%** |
| no-hardhat AP50 (test) | **96.6%** |

All five classes, including the minority violation classes (vest, no-vest), are above 91% AP50 on the held-out test set. This is well above the threshold needed for a reliable deployment.

### 2. Compute constraints

- **RT-DETR-l** has ~32M parameters (3.4x YOLO11s, 12x YOLO11n)
- At  on an RTX 4060 Laptop (8.6 GB VRAM), estimated training time is **40-50 hours**
- The expected marginal gain at 93%+ mAP50 does not justify the compute cost

### 3. Time constraints

The project timeline requires moving to evaluation (NB07) and deployment (NB08). The two-model comparison (YOLO11n vs YOLO11s) already provides:
- A meaningful speed vs accuracy tradeoff analysis
- Two deployment options: YOLO11n for CPU/edge (fast), YOLO11s for GPU (accurate)
